```python
from data_import import *
import pandas as pf
from datetime import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_weights = pd.read_csv('all_weights.csv',index_col=0)
df_weights
```

```python
df_data = pd.read_csv('all_data.csv',index_col=0)
df_data
```

```python
def Train_get_data(quarter_number,df_weights,df_data):
    df = df_weights[df_weights['Time']==quarter_number].reset_index(drop=True)
    return df_data[df['code']][(df_data.index>=df['Train_Start'][0])&(df_data.index<=df['Train_End'][0])].dropna()
Train_df_2016q2 = Train_get_data('2016q2',df_weights,df_data).sort_index(ascending=True)
Train_df_2016q2
```

```python
def get_data(quarter_number,df_weights,df_data):
    df = df_weights[df_weights['Time']==quarter_number].reset_index(drop=True)
    return df_data[df['code']][(df_data.index>=df['Start_Date'][0])&(df_data.index<=df['End_Date'][0])].dropna()
df_2016q2 = get_data('2016q2',df_weights,df_data).sort_index(ascending=True)
df_2016q2
```

```python
def get_index(quarter_number,df_weights,df_data):
    df = df_weights[df_weights['Time']==quarter_number].reset_index(drop=True)
    return df_data['Price'][(df_data.index>=df['Start_Date'][0])&(df_data.index<=df['End_Date'][0])].dropna()
```

```python
#Define Factor Weights
dic_factor_weights = {'a50':0.5,'mom':0.25,'size':0.25}
def get_return(quarter_name,df_weights,df_data,mom_number_comp,size_number_comp,dic_factor_weights):
    df_train = Train_get_data(quarter_name,df_weights,df_data).sort_index(ascending=True)
    df_test = get_data(quarter_name,df_weights,df_data).sort_index(ascending=True)
    #get mom indicator
    df_mom = df_train.apply(lambda x:(x[-1]-x[0])/x[0]).sort_values(ascending=False).head(mom_number_comp)
    #get components
    df_quarter_weights = df_weights[['code','weight']][df_weights['Time']==quarter_name].set_index('code')
    #construct ETF
    df_etf = pd.DataFrame()
    #Calculate A50
    df_a50 = pd.DataFrame()
    for name in df_quarter_weights.index:
        df_a50[name] = df_test[name].apply(lambda x: x*df_quarter_weights.loc[name])
    df_etf['a50'] = df_a50.sum(axis=1)
    #Calculate Mom
    mom = pd.DataFrame()
    mom_weight = 1/mom_number_comp
    for name in df_mom.index:
        mom[name] = df_test[name].apply(lambda x: x*mom_weight)
    df_etf['mom'] = mom.sum(axis=1)
    #Calculate Size
    df_size = pd.DataFrame()
    for name in df_quarter_weights.head(size_number_comp).index:
        df_size[name] = df_test[name].apply(lambda x: x*df_quarter_weights.loc[name])
    df_etf['size'] = df_size.sum(axis=1)
    #Calculate portfolio
    df_etf['portfolio'] = dic_factor_weights['a50']*df_etf['a50']+dic_factor_weights['mom']*df_etf['mom']+dic_factor_weights['size']*df_etf['size']
    df_etf['Returns'] = np.log(df_etf['a50']/df_etf['a50'].shift(1))
    df_etf['Strategy_Returns'] = np.log(df_etf['portfolio']/df_etf['portfolio'].shift(1))
    df_etf['Cum_Ret'] = df_etf['Returns'].cumsum().apply(np.exp)
    df_etf['Cum_Str'] = df_etf['Strategy_Returns'].cumsum().apply(np.exp)
    
    return df_etf
```

```python
mom_number_comp = 10
size_number_comp = 10
df = get_return('2016q2',df_weights,df_data,mom_number_comp,size_number_comp,dic_factor_weights)
df[['Cum_Ret','Cum_Str']].plot(figsize=[18,8])
```

```python
df
```

```python
dict_all = {}
for name in df_weights['Time'].sort_values(ascending=True).unique()[1:]:
    df = get_return(name,df_weights,df_data,mom_number_comp,size_number_comp,dic_factor_weights)
    dict_all[name] = df
    print(name)
    print(df)
    df[['Cum_Ret','Cum_Str']].plot(figsize=[18,8],title=name)
    plt.show()
```

```python
dict_summary = {'A50_Return':[],'Strategy_Return':[],'Index_Return':[]}
for name in list(dict_all.keys()):
    dict_summary['A50_Return'].append(dict_all[name]['Cum_Ret'][-1])
    dict_summary['Strategy_Return'].append(dict_all[name]['Cum_Str'][-1])
    df_index = pd.DataFrame()
    df_index['Price'] = get_index(name,df_weights,df_data).sort_index(ascending=True)
    df_index['Returns'] = np.log(df_index['Price']/df_index['Price'].shift(1))
    df_index['Cum_Ret'] = df_index['Returns'].cumsum().apply(np.exp)
    dict_summary['Index_Return'].append(df_index['Cum_Ret'][-1])
df_summary = pd.DataFrame(dict_summary).set_index(df_weights['Time'].sort_values(ascending=True).unique()[1:])
```
